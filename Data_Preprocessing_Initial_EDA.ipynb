## Outlier Detection

```python
target_cols = ['actual_time', 'osrm_time', 'actual_distance_to_destination', 'segment_actual_time']


def detect_outliers_iqr(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound
```

```python
# Apply and plot for each column

for col in target_cols:
    if col in df.columns:
        outliers, lb, ub = detect_outliers_iqr(col)
        print(f"\n Column: {col}")
        print(f"- Outliers detected: {outliers.shape[0]}")
        print(f"- Lower Bound: {lb:.2f}, Upper Bound: {ub:.2f}")

        # Visualize with boxplot
        
        plt.figure(figsize=(6, 1.5))
        sns.boxplot(x=df[col], color='skyblue')
```
